My attempt at two player Tic Tac Toe game in Python - part of the https://www.practicepython.org/ beginner Python series

```python
def print_board(board): # function that prints the board to look like a 3x3 grid
    print(board[1] + ' | ' + board[2] + '| ' + board[3])
    print('--+--+--')
    print(board[4] + ' | ' + board[5] + '| ' + board[6])
    print('--+--+--')
    print(board[7] + ' | ' + board[8] + '| ' + board[9])

def check_winner(board): #function that checks to see if there is a winner

    if board[1] == board[2] == board[3] != ' ': #row 1 winner
        return True
    elif board[4] == board[5] == board[6] != ' ': #row 2 winner
        return True
    elif board[7] == board[8] == board[9] != ' ': #row 3 winner
        return True
    elif board[1] == board[4] == board[7] != ' ': #column 1 winner
        return True
    elif board[2] == board[5] == board[8] != ' ': #column 2 winner
        return True
    elif board[3] == board[6] == board[9] != ' ': #column 3 winner
        return True
    elif board[1] == board[5] == board[9] != ' ': #diagonal 1 winner
        return True
    elif board[3] == board[5] == board[7] != ' ': #diagonal 2 winner
        return True
    else:
        return False 

def gameplay(board): #function that programs the rules of the game and user moves
    turn = 'X' #assume X goes first
    count = 0  #number of moves

    while count < 10:

        print_board(board)
        print("\n It's your turn", turn, "\n")
        move = input("Enter a number between 1 and 9 to make your move: ")

        if move.isdigit() == True and int(move) in range(1,10) and board[int(move)] == ' ': #criteria for a valid entry
            board[int(move)] = turn #update the appropriate square on the board with X or O
            count += 1
        else:
            print("\n Invalid move. Please try again")
            continue

        if check_winner(board) == True: #call check_winner function to see if the game was won on that turn
            print_board(board)
            print("\nGame over.", turn, "wins.")
            break
        
        elif count == 9: #if nobody's won after all squares are full, then it's a tie
            print_board(board)
            print("\nGame over. It's a tie.")
            break

        if turn == 'X': #alternates moves between X and O
            turn = 'O'
        else:
            turn = 'X' 
          
    restart = input("\nWould you like to play again? (Y/N).") #option to play again once the game ends  
    if restart == 'y' or restart == 'Y':
        board = {1:' ', 2:' ', 3:' ', #resetting the board
        4:' ', 5:' ', 6:' ', 
        7:' ', 8:' ', 9:' '}

        gameplay(board) #starts the gameplay flow again
    else:
        print("\nThanks for playing. Goodbye.")

print("Welcome to two-player Tic Tac Toe. \n") 
print("Each square on the board corresponds to a number between 1 and 9. The top left square is number 1 and the bottom right square is number 9.")
print("You will make moves in the game by specifying the number of the square you want to play on.\n X goes first.\n")

board = {1:' ', 2:' ', 3:' ', 
        4:' ', 5:' ', 6:' ', 
        7:' ', 8:' ', 9:' '} #starting blank board as a dictionary

if __name__ == "__main__":
    gameplay(board)
```
